```python
import numpy as np
import pandas as pd
import re
import ast
```

```python
df = pd.read_csv(r"Dataset\data\dresses_bd_processed_data.csv")
```

```python
def clean_text(s):
    s = str(s).lower()
    s = re.sub(r"<[^>]+>", " ", s)       
    s = re.sub(r"http\S+", " ", s)       
    s = re.sub(r"[^a-z0-9 ,\.]", " ", s) 
    return " ".join(s.split())
```

```python
for col in ["meta_info", "product_name", "description"]:
    df[col] = df[col].fillna("").apply(clean_text)
```

```python
def clean_brackets(df, columns):
    """
    Removes square brackets [] and curly braces {} from string values
    in the specified columns of a DataFrame.

    Args:
        df (pd.DataFrame): The DataFrame to clean.
        columns (list): List of column names to process.
    """
    for col in columns:
        if col in df.columns:
            df[col] = df[col].apply(
                lambda x: (
                    x.replace('[', '')
                     .replace(']', '')
                     .replace('{', '')
                     .replace('}', '')
                ) if isinstance(x, str) else x
            )
    return df
```

```python
rmv_str_df = clean_brackets(df ,["feature_list","style_attributes",])
```

```python
def replace_nan(df):
    """
    Replaces all NaN values in a DataFrame with empty strings.
    """
    return df.replace(np.nan, '', regex=False)
```

```python
all_tags = [sublist for sublist in df['feature_list']]
all_tags += [sublist for sublist in df['style_attributes']]
tag_freq = Counter(all_tags)

MIN_FREQ = 5
MAX_FREQ = 0.8 * len(df)

filtered_tags = {
    tag for tag, count in tag_freq.items() 
    if MIN_FREQ <= count <= MAX_FREQ
}

print(f"Original vocabulary: {len(tag_freq)}")
print(f"Filtered vocabulary: {len(filtered_tags)}")
```
